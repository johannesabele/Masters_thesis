# Visualizing Neuroimages in nii format

```python
import os
import numpy as np
# Use for example data included in nibabel package
#from nibabel.testing import data_path  
import nibabel as nib
```

```python
filename = "sub-001000_T1w.nii.gz"
#filename = os.path.join(data_path, 'example4d.nii.gz')
```

```python
img = nib.load(filename)
img.shape
#img.affine.shape
```

```python
data = img.get_fdata()
data.shape
hdr = img.header
print(hdr)
```

```python
# Visualize image
import matplotlib.pyplot as plt

def show_slices(slices):
   """ Function to display row of image slices """
   fig, axes = plt.subplots(1, len(slices))
   for i, slice in enumerate(slices):
       axes[i].imshow(slice.T, cmap="gray", origin="lower")

# Transform data 
epi_img_data = img.get_fdata()
print("Image shape: ", epi_img_data.shape)

slice_0 = epi_img_data[80, :, :]
slice_1 = epi_img_data[:, 112, :]
slice_2 = epi_img_data[:, :, 80]

print("Type img data: ", type(epi_img_data))
print(max(slice_0[slice_0 > 0]))

show_slices([slice_0, slice_1, slice_2])

plt.suptitle("Center slices for EPI image")  
```
